```python
import numpy as np
import matplotlib.pyplot as plt

class MLP:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size

        # initialize weights matrix and biases
        self.W_input_hidden = np.random.randn(self.input_size, self.hidden_size)
        self.b_input_hidden = np.zeros((1, self.hidden_size))
        self.W_hidden_output = np.random.randn(self.hidden_size, self.output_size)
        self.b_hidden_output = np.zeros((1, self.output_size))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def d_sigmoid(self, x):
        return x * (1 - x)

    def forward(self, input_data):
        hidden_layer_input = np.dot(input_data, self.W_input_hidden) + self.b_input_hidden
        hidden_layer_output = self.sigmoid(hidden_layer_input)
        output_layer_input = np.dot(hidden_layer_output, self.W_hidden_output) + self.b_hidden_output
        output = self.sigmoid(output_layer_input)
        return hidden_layer_output, output

    def backward(self, input_data, target, hidden_output, output, lr=0.1):
        output_error = target - output
        output_grad = output_error * self.d_sigmoid(output)
        hidden_error = np.dot(output_grad, self.W_hidden_output.T)
        hidden_grad = hidden_error * self.d_sigmoid(hidden_output)
        
        self.W_hidden_output += np.dot(hidden_output.T, output_grad) * lr
        self.b_hidden_output += np.sum(output_grad, axis=0, keepdims=True) * lr
        self.W_input_hidden += np.dot(input_data.T, hidden_grad) * lr
        self.b_input_hidden += np.sum(hidden_grad, axis=0, keepdims=True) * lr

    def train(self, input_data, target, epochs=1000):
        for epoch in range(epochs):
            for x, y in zip(input_data, target):
                x = x.reshape(1, -1)
                y = y.reshape(1, -1)
                hidden_output, output = self.forward(x)
                self.backward(x, y, hidden_output, output)

    def predict(self, input_data):
        _, output = self.forward(input_data)
        return output

# XOR truth table values
xor_input = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
xor_target = np.array([[0], [1], [1], [0]])

# MLP parameters
input_size = 2
hidden_size = 10
output_size = 1

# Instantiate the MLP
model = MLP(input_size, hidden_size, output_size)
model.train(xor_input, xor_target, epochs=10000)

# Test inputs
test_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
predictions = model.predict(test_inputs)

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot of inputs
for i, input_data in enumerate(test_inputs):
    if predictions[i] >= 0.5:
        plt.scatter(input_data[0], input_data[1], color='r')
    else:
        plt.scatter(input_data[0], input_data[1], color='b')

plt.title("MLP Classification Results for XOR Problem")
plt.xlabel("Input Feature 1")
plt.ylabel("Input Feature 2")
plt.legend(["Class 1", "Class 0"])
plt.grid(True)
plt.show()

# Print predictions for each input
print("Predicted 00 output:", predictions[0])
print("Predicted 01 output:", predictions[1])
print("Predicted 10 output:", predictions[2])
print("Predicted 11 output:", predictions[3])
```

```python
import numpy as np
import matplotlib.pyplot as plt

class MLP:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size

        # Initialize weights matrix and biases
        self.W_input_hidden = np.random.randn(self.input_size, self.hidden_size)
        self.b_input_hidden = np.zeros((1, self.hidden_size))
        self.W_hidden_output = np.random.randn(self.hidden_size, self.output_size)
        self.b_hidden_output = np.zeros((1, self.output_size))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def d_sigmoid(self, x):
        return x * (1 - x)
    
    def relu(self, x):
        return np.maximum(0, x)

    def d_relu(self, x):
        return np.where(x > 0, 1, 0)

    def forward(self, input_data):
        hidden_layer_input = np.dot(input_data, self.W_input_hidden) + self.b_input_hidden
        hidden_layer_output = self.relu(hidden_layer_input)
        output_layer_input = np.dot(hidden_layer_output, self.W_hidden_output) + self.b_hidden_output
        output = self.sigmoid(output_layer_input)
        return hidden_layer_output, output

    def backward(self, input_data, target, hidden_output, output, lr=0.01):
        output_error = target - output
        output_grad = output_error * self.d_sigmoid(output)
        hidden_error = np.dot(output_grad, self.W_hidden_output.T)
        hidden_grad = hidden_error * self.d_relu(hidden_output)
        
        self.W_hidden_output += np.dot(hidden_output.T, output_grad) * lr
        self.b_hidden_output += np.sum(output_grad, axis=0, keepdims=True) * lr
        self.W_input_hidden += np.dot(input_data.T, hidden_grad) * lr
        self.b_input_hidden += np.sum(hidden_grad, axis=0, keepdims=True) * lr

    def train(self, input_data, target, epochs=100):
        for epoch in range(epochs):
            epoch_loss = 0
            for x, y in zip(input_data, target):
                x = x.reshape(1, -1)
                y = y.reshape(1, -1)
                hidden_output, output = self.forward(x)
                self.backward(x, y, hidden_output, output)
                epoch_loss += np.mean(np.square(y - output))
            if epoch % 1000 == 0:
                print(f"Epoch {epoch}, Loss: {epoch_loss/len(input_data)}")

    def predict(self, input_data):
        _, output = self.forward(input_data)
        return output

# Generate data
X = np.concatenate([np.random.random((50, 2)) * .9 + np.array([1, 1]), np.random.random((50, 2)) * .9 + np.array([2, 2])])
Y = np.concatenate([np.ones((50, 1)), np.zeros((50, 1))])  # Corrected labels to 0 and 1

# Normalize data
X = (X - np.min(X, axis=0)) / (np.max(X, axis=0) - np.min(X, axis=0))

# MLP parameters
input_size = 2
hidden_size = 20
output_size = 1

# Instantiate the MLP
model = MLP(input_size, hidden_size, output_size)
model.train(X, Y, epochs=100)

# Predict
predictions = model.predict(X)

# Convert predictions to class labels
predicted_classes = predictions >= 0.5

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot of inputs with predicted classes
for i, input_data in enumerate(X):
    if predicted_classes[i]:
        plt.scatter(input_data[0], input_data[1], color='r', label='Class 1' if i == 0 else "")
    else:
        plt.scatter(input_data[0], input_data[1], color='b', label='Class 0' if i == 0 else "")


plt.title("MLP Classification Results")
plt.xlabel("Input Feature 1")
plt.ylabel("Input Feature 2")
plt.legend()
plt.grid(True)
plt.show()

# Print some predictions for verification
for i in range(5):
    print(f"Input: {X[i]}, Predicted output: {predictions[i]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os

class MLP:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size

        # Initialize weights matrix and biases
        self.W_input_hidden = np.random.randn(self.input_size, self.hidden_size)
        self.b_input_hidden = np.zeros((1, self.hidden_size))
        self.W_hidden_output = np.random.randn(self.hidden_size, self.output_size)
        self.b_hidden_output = np.zeros((1, self.output_size))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def d_sigmoid(self, x):
        return x * (1 - x)

    def forward(self, input_data):
        hidden_layer_input = np.dot(input_data, self.W_input_hidden) + self.b_input_hidden
        hidden_layer_output = self.sigmoid(hidden_layer_input)
        output_layer_input = np.dot(hidden_layer_output, self.W_hidden_output) + self.b_hidden_output
        output = self.sigmoid(output_layer_input)
        return hidden_layer_output, output

    def backward(self, input_data, target, hidden_output, output, lr=0.01):
        output_error = target - output
        output_grad = output_error * self.d_sigmoid(output)
        hidden_error = np.dot(output_grad, self.W_hidden_output.T)
        hidden_grad = hidden_error * self.d_sigmoid(hidden_output)
        
        self.W_hidden_output += np.dot(hidden_output.T, output_grad) * lr
        self.b_hidden_output += np.sum(output_grad, axis=0, keepdims=True) * lr
        self.W_input_hidden += np.dot(input_data.T, hidden_grad) * lr
        self.b_input_hidden += np.sum(hidden_grad, axis=0, keepdims=True) * lr

    def train(self, input_data, target, epochs=1000):
        self.losses = []
        for epoch in range(epochs):
            epoch_loss = 0
            for x, y in zip(input_data, target):
                x = x.reshape(1, -1)
                y = y.reshape(1, -1)
                hidden_output, output = self.forward(x)
                self.backward(x, y, hidden_output, output)
                epoch_loss += np.mean(np.square(y - output))
            self.losses.append(epoch_loss / len(input_data))
            if epoch % 10 == 0:
                print(f"Epoch {epoch}, Loss: {self.losses[-1]}")

    def predict(self, input_data):
        _, output = self.forward(input_data)
        return output

# Function to load and preprocess images
def load_images_from_folder(folder, label, image_size=(64, 64)):
    images = []
    labels = []
    for filename in os.listdir(folder):
        if filename.endswith(".jpg"):
            img_path = os.path.join(folder, filename)
            img = Image.open(img_path).convert('L')
            img = img.resize(image_size)
            img = np.array(img).flatten()
            img = img / 255.0  # Normalize pixel values
            images.append(img)
            labels.append(label)
    return images, labels

# Load datasets
piano_images, piano_labels = load_images_from_folder('dataset/piano', [1, 0, 0])
clavier_images, clavier_labels = load_images_from_folder('dataset/clavier', [0, 1, 0])
machine_a_ecrire_images, machine_a_ecrire_labels = load_images_from_folder('dataset/machine_a_ecrire', [0, 0, 1])

# Combine datasets
X = np.array(piano_images + clavier_images + machine_a_ecrire_images)
Y = np.array(piano_labels + clavier_labels + machine_a_ecrire_labels)

# Shuffle the dataset
indices = np.arange(X.shape[0])
np.random.shuffle(indices)
X = X[indices]
Y = Y[indices]

# MLP parameters
input_size = X.shape[1]  # Number of pixels in the flattened image
hidden_size = 100  # You can adjust this number
output_size = 3  # Number of classes

# Instantiate the MLP
model = MLP(input_size, hidden_size, output_size)
model.train(X, Y, epochs=50)

# Predict
predictions = model.predict(X)

# Convert predictions to class labels
predicted_classes = np.argmax(predictions, axis=1)
true_classes = np.argmax(Y, axis=1)

# Calculate accuracy
accuracy = np.mean(predicted_classes == true_classes)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Plot the training loss
plt.figure(figsize=(10, 6))
plt.plot(model.losses)
plt.title("Training Loss over Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

# Plot some test images with their predicted and true labels
for i in range(5):
    img = X[i].reshape((64, 64))
    plt.imshow(img, cmap='gray')
    plt.title(f"True: {['piano', 'clavier', 'machine_a_ecrire'][true_classes[i]]}, Pred: {['piano', 'clavier', 'machine_a_ecrire'][predicted_classes[i]]}")
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

class MLP:
    def __init__(self, input_size, hidden_layers, hidden_size, output_size, lr=0.001):
        self.input_size = input_size
        self.hidden_layers = hidden_layers  # Nombre de couches cachées
        self.hidden_size = hidden_size  # Nombre de neurones par couche cachée
        self.output_size = output_size
        self.lr = lr  # Learning rate

        
        self.weights = []
        self.biases = []

        self.weights.append(np.random.randn(self.input_size, self.hidden_size))
        self.biases.append(np.zeros((1, self.hidden_size)))

        for _ in range(self.hidden_layers - 1):
            self.weights.append(np.random.randn(self.hidden_size, self.hidden_size))
            self.biases.append(np.zeros((1, self.hidden_size)))

     
        self.weights.append(np.random.randn(self.hidden_size, self.output_size))
        self.biases.append(np.zeros((1, self.output_size)))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def d_sigmoid(self, x):
        return x * (1 - x)

    def forward(self, input_data):
        activations = [input_data]
        for i in range(self.hidden_layers):
            hidden_input = np.dot(activations[-1], self.weights[i]) + self.biases[i]
            hidden_output = self.sigmoid(hidden_input)
            activations.append(hidden_output)

        # Dernière couche
        output_input = np.dot(activations[-1], self.weights[-1]) + self.biases[-1]
        output = self.sigmoid(output_input)
        activations.append(output)

        return activations

    def backward(self, input_data, target, activations):
        output = activations[-1]
        output_error = target - output
        output_grad = output_error * self.d_sigmoid(output)

        # Mise à jour des poids et biais pour la couche de sortie
        self.weights[-1] += np.dot(activations[-2].T, output_grad) * self.lr
        self.biases[-1] += np.sum(output_grad, axis=0, keepdims=True) * self.lr

        # Propagation des erreurs et mise à jour des couches cachées
        for i in reversed(range(self.hidden_layers)):
            hidden_error = np.dot(output_grad, self.weights[i + 1].T)
            hidden_grad = hidden_error * self.d_sigmoid(activations[i + 1])
            self.weights[i] += np.dot(activations[i].T, hidden_grad) * self.lr
            self.biases[i] += np.sum(hidden_grad, axis=0, keepdims=True) * self.lr
            output_grad = hidden_grad

    def train(self, input_data, target, epochs=100):
        self.losses = []
        for epoch in range(epochs):
            epoch_loss = 0
            for x, y in zip(input_data, target):
                x = x.reshape(1, -1)
                y = y.reshape(1, -1)
                activations = self.forward(x)
                self.backward(x, y, activations)
                epoch_loss += np.mean(np.square(y - activations[-1]))
            self.losses.append(epoch_loss / len(input_data))
            print(f"Epoch {epoch}, Loss: {self.losses[-1]}")

    def predict(self, input_data):
        return np.argmax(self.forward(input_data)[-1], axis=1)

    def save_weights(self, file_path):
    # Préparer un dictionnaire pour sauvegarder les poids et biais
        weights_dict = {}
        biases_dict = {}

    # Sauvegarder chaque couche de poids avec une clé unique
        for i, (w, b) in enumerate(zip(self.weights, self.biases)):
            weights_dict[f'weight_{i}'] = np.asarray(w)
            biases_dict[f'bias_{i}'] = np.asarray(b)

    # Sauvegarder dans un fichier .npz
        np.savez(file_path, **weights_dict, **biases_dict)
        print(f"Weights and biases saved to {file_path}")




    def load_weights(self, file_path):
        data = np.load(file_path, allow_pickle=True)
        self.weights = data['weights']
        self.biases = data['biases']
        print(f"Weights loaded from {file_path}")


def load_images_from_folder(folder, label, image_size=(64, 64)):
    images = []
    labels = []
    for filename in os.listdir(folder):
        if filename.endswith(".jpg"):
            img_path = os.path.join(folder, filename)
            img = Image.open(img_path).convert('L')
            img = img.resize(image_size)
            img = np.array(img).flatten()
            img = img / 255.0  # Normaliser les valeurs des pixels
            images.append(img)
            labels.append(label)
    return images, labels


piano_images, piano_labels = load_images_from_folder('dataset/piano', [1, 0, 0])
clavier_images, clavier_labels = load_images_from_folder('dataset/clavier', [0, 1, 0])
machine_a_ecrire_images, machine_a_ecrire_labels = load_images_from_folder('dataset/machine_a_ecrire', [0, 0, 1])


X = np.array(piano_images + clavier_images + machine_a_ecrire_images)
Y = np.array(piano_labels + clavier_labels + machine_a_ecrire_labels)


indices = np.arange(X.shape[0])
np.random.shuffle(indices)
X = X[indices]
Y = Y[indices]


split_ratio = 0.8
split_index = int(len(X) * split_ratio)
X_train, X_test = X[:split_index], X[split_index:]
Y_train, Y_test = Y[:split_index], Y[split_index:]

# Paramètres de MLP
input_size = X.shape[1]  # Nombre de pixels dans l'image aplatie
hidden_layers = 3  # Nombre de couches cachées
hidden_size = 150  # Nombre de neurones par couche cachée
output_size = 3  # Nombre de classes (piano, clavier, machine à écrire)
learning_rate = 0.001  # Taux d'apprentissage
epochs = 280  # Nombre d'épochs


model = MLP(input_size, hidden_layers, hidden_size, output_size, lr=learning_rate)
model.train(X_train, Y_train, epochs=epochs)


model.save_weights('C:/Users/nabil/Desktop/MLP/mlp_weights_V3.npz')


predictions = model.predict(X_test)


true_classes = np.argmax(Y_test, axis=1)
accuracy = np.mean(predictions == true_classes)
print(f"Accuracy: {accuracy * 100:.2f}%")


plt.figure(figsize=(10, 6))
plt.plot(model.losses)
plt.title("Training Loss over Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid(True)
plt.show()


conf_matrix = confusion_matrix(true_classes, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=['piano', 'clavier', 'machine_a_ecrire'])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.show()

```
